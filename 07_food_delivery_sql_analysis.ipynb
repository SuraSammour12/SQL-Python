```python
# Import Libraries
import pandas as pd
import sqlite3
```

```python
# Load CSV file
df = pd.read_csv('food_delivery_data_large.csv')
df
```

```python
# Create SQLite DB + table
conn = sqlite3.connect('food_delivery.db')
df.to_sql('restaurants',conn,if_exists='replace',index=False)
```

Querying with SQL + Pandas

```python
# select
pd.read_sql_query('select * from restaurants',conn)
```

#  Filtering with WHERE

```python
# Display only restaurants that serve "Indian" cuisine:
query = "select * from restaurants where cuisine = 'Indian';"
df_indian = pd.read_sql_query(query, conn)
df_indian.head()
```

```python
# Restaurants with a rating higher than 4.5:
query = "select name, city, rating from restaurants where rating > 4.5;"
df_top_rated = pd.read_sql_query(query, conn)
df_top_rated
```

#  Sorting with ORDER BY

```python
# Sort restaurants by rating in descending order:
query = "select name, rating from restaurants order by rating desc;"
df_sorted_rating = pd.read_sql_query(query, conn)
df_sorted_rating.head()
```

```python
# Sort by city in ascending order, and within each city, rating in descending order:
query = """
select name, city, rating
from restaurants
order by city asc, rating desc;
"""
df_sorted_city_rating = pd.read_sql_query(query, conn)
df_sorted_city_rating.head()
```

# Aggregation: avg, count

```python
query="select city, avg(average_cost) as avg_cost from restaurants group by city"
pd.read_sql_query(query,conn)
```

# Visualization with Matplotlib / Seaborn


# Plot ratings vs cost

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt


# Plot ratings vs cost
sns.scatterplot(data=df, x="average_cost", y="rating", hue="city")
plt.title("Rating vs Average Cost")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()

```

# Distribution of cuisines

```python
# Cuisine distribution
sns.countplot(data=df, x="cuisine")
plt.xticks(rotation=45)
plt.title("Cuisine Distribution")
plt.show()
```

#  Visualization: Top Rated Restaurants

```python
query = """
select name, city, rating
from restaurants
order by rating desc
limit 10;
"""
top_rated = pd.read_sql_query(query, conn)
top_rated
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.barplot(data=top_rated, x="rating", y="name", hue="city", dodge=False)
plt.title("Top 10 Highest Rated Restaurants")
plt.xlabel("Rating")
plt.ylabel("Restaurant Name")
plt.legend(title="City", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
