```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('books.db')
cursor = conn.cursor()
```

```python
cursor.execute("""
create table if not exists books(
  book_id text primary key,
  title text,
  author text,
  year integer
)
""")
```

```python
cursor.execute("insert into books values ('B1', '1984', 'George Orwell', 1949)")
cursor.execute("insert into books values ('B2', 'To Kill a Mockingbird', 'Harper Lee', 1960)")
cursor.execute("insert into books values ('B3', 'Pride and Prejudice', 'Jane Austen', 1813)")
cursor.execute("insert into books values ('B4', 'The Great Gatsby', 'F. Scott Fitzgerald', 1925)")
cursor.execute("insert into books values ('B5', 'The Catcher in the Rye', 'J.D. Salinger', 1951)")
cursor.execute("insert into books values ('B6', 'The Hobbit', 'J.R.R. Tolkien', 1937)")
cursor.execute("insert into books values ('B7', 'Fahrenheit 451', 'Ray Bradbury', 1953)")
cursor.execute("insert into books values ('B8', 'Jane Eyre', 'Charlotte Brontë', 1847)")
cursor.execute("insert into books values ('B9', 'Moby-Dick', 'Herman Melville', 1851)")
cursor.execute("insert into books values ('B10', 'The Alchemist', 'Paulo Coelho', 1988)")

```

```python
conn.commit()
```

SELECT * – Retrieve All Data from a Table

```python
df = pd.read_sql_query('select * from books',conn)
df
```

Selecting Specific Columns from a Table

```python
df1 = pd.read_sql_query('select title,author from books;',conn)
df1
```

Filtering Rows with WHERE

```python
# Get all books published after 1950
df  =pd.read_sql_query('select * from books where year>1950',conn)
df
```

```python
# Get books written by George Orwell
df = pd.read_sql_query("select title,year from books where author='George Orwell';",conn)
```

```python
df
```

```python
# Get the book with ID: B3
df = pd.read_sql_query('select * from books where book_id="B3";',conn)
df
```

 Comparison Operators in SQL

```python
# Books published before 1950
df = pd.read_sql_query('select title,year from books where year < 1950',conn)
df
```

```python
# Books not written by George Orwell
df = pd.read_sql_query('select title from books where author!="George Orwell"',conn)
df
```

```python
# Books published in or after 1960
df = pd.read_sql_query('select title from books where year>=1960',conn)
df
```

Sorting Results with ORDER BY | ASC (default), DESC

```python
# Sort books from oldest to newest
df = pd.read_sql_query('select title,year from books order by year asc;',conn)
df
```

```python
# Sort books by title alphabetically
df = pd.read_sql_query('select title from books order by title;',conn)
df
```

```python
# Sort books from newest to oldest
df = pd.read_sql_query('select title,year from books order by year desc;',conn)
df
```

sort by multiple columns

```python
df = pd.read_sql_query('select title, author,year from books order by author asc,year desc;',conn)
df
```

Using DISTINCT and LIMIT
* DISTINCT: Used to return unique values only (remove duplicates)
* LIMIT: Used to limit the number of rows returned

```python
# DISTINCT
df = pd.read_sql_query('select distinct author from books;',conn)
df
```

```python
# LIMIT
df = pd.read_sql_query('select title, year from books limit 3;',conn)
df
```

```python
# Combine with sorting
df = pd.read_sql_query('select title,year from books order by year desc limit 5;',conn)
df
```
